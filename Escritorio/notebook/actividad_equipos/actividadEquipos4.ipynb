# Constructor Inmobiliario

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_csv("data/ConstructorInmobiliario.csv")
```

```python
X = df[['Income', 'Size', 'Senior', 'Education']]
X.head()
```

```python
y = df[['SqrFt']]
y.head()
```

```python
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
y_pred
```

### Variables

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Error cuadrático de la media: {mse}")
print(f"Varianza: {r2}")
```

```python
model.intercept_
```

```python
print(f"Coeficientes: {model.coef_}")
```
