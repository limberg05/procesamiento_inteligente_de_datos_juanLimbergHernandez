### Regresión Lineal Múltiple - SalsBerry Realty
1. Determinar las variables independientes
2. Determinar la variable dependiente
3. Obtener el valor de interseccion
4. Estimar el corto de calefaccion para una Temperatura externa media de 30 grados, 5 pulgadas de aislamiento y un calentador con antiguedad de 10 años.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Importar datos
df = pd.read_csv('data/SalsberryRealty.csv')
df.head()
```

# Determinar las variables independienetes
x1, x2, y3

```python
X = df[['Temp', 'Insul', 'Age']]
X
```

```python
# 2. Determinar las variables Dependientes
```

```python
y = df[['Cost']]
y
```

```python
# Ajustar el modelo de regresión multiple
model = LinearRegression()
model.fit(X,y) # entrenando el modelo
y_pred = model.predict(X)
#y_pred

# Métricas estadísticas
mse = mean_squared_error(y, y_pred) # Error de la media cuadrática
r2 = r2_score(y,y_pred) # varianza
```

# Calcular el valor de Interseccion

```python
model.intercept
```

```python
# Coeficiente
print("Coeficientes: ", model.coef_)
print("Error cuadrático medio",mse)
print("Varianza: ", r2)
```

```python
Y = model.predict([[30,5,10]])
print("El costo de calefaccion estimado es: ", Y[0])
```
