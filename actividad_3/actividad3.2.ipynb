```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df_diabetes = datasets.load_diabetes()

x = df_diabetes.data[:, 2].reshape(-1, 1)
y = df_diabetes.target

X_train = x[:-20]
X_test  = x[-20:]

Y_train = y[:-20]
Y_test  = y[-20:]

regressor = LinearRegression()
regressor.fit(X_train, Y_train)

y_pred = regressor.predict(X_test)

print("coeficients:\n", regressor.coef_)
print("indepentent term:\n", regressor.intercept_)
print("mean square error: %.2f" % mean_squared_error(Y_test, y_pred))
print("r2 score: %.2f" % r2_score(Y_test, y_pred))

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].scatter(X_train, Y_train)
ax[0].plot(X_train, regressor.predict(X_train), c="g")
ax[0].set_title("linear regression (train)")
ax[0].set_ylabel("target")

ax[1].scatter(X_test, Y_test)
ax[1].plot(X_test, y_pred, c="g")
ax[1].set_title("linear regression (test)")
ax[1].set_xlabel("BMI")
ax[1].set_ylabel("target")

plt.suptitle("linear regression model")
plt.show()
```

El modelo de regresin lineal muestra que existe una relacin entre el ndice de masa corporal y la progresin de la diabetes. Al separar los datos de forma determinista, se pudo evaluar mejor el desempeo del modelo. Aunque el error indica que las predicciones no son totalmente precisas, el valor de R^2 confirma que el BMI es un predictor relevante, aunque por s solo no explica todo el comportamiento del modelo.
