# Regresión Lineal Múltiple - SalsBerry Realty

1. Determinal las variables independientes
2. Determinar la variable dependiente
3. Obtener el valor de intesección
4. Estimar el corto de calefacción para una Temperatura externa media de 30°, 5 pulfadas de aislamiento y un calentador con antigüedad de 10 años.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("data/SalsberryRealty.csv")
df.head()
```

1. Determinal las variables independientes X



```python
X = df[['Temp', 'Insul','Age']]
```

2. Determinar la variable dependiente Y


```python
y = df[['Cost']]
```

ajustar el modelo a regresion multiple


```python
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)
```

Metricas Estadisticas

```python
mse = mean_squared_error(y,y_pred)
r2 = r2_score(y, y_pred)
```

3. Obtener el valor de intesección


```python
model.intercept_
```

```python
#coeficientes
print('Coeficientes: ', model.coef_)
print('Error cuadratico medio',mse)
print('Varianza: ', r2)
```

estimar para 30°, asilamiento 5", atiguedad 10 años

```python
Y = model.predict([[30,5,10]])
print("el costo de calefaccion estimado es: ", Y)
```
