```python
import random

class Individuo:
    def __init__(self):
        self.cromosoma = [random.randint(0, 1) for _ in range(18)]
        self.aptitud = 0

    def evaluar_aptitud(self):
        self.aptitud = sum(self.cromosoma)

    def mutar(self):
        for i in range(len(self.cromosoma)):
            if random.random() < 0.05:
                self.cromosoma[i] = 1 - self.cromosoma[i]

    def cruzar(self, otro):
        punto = random.randint(1, len(self.cromosoma) - 1)
        hijo = Individuo()
        hijo.cromosoma = self.cromosoma[:punto] + otro.cromosoma[punto:]
        return hijo

    def mostrar(self):
        print("Cromosoma:", self.cromosoma)
        print("Aptitud:  ", self.aptitud)


class Poblacion:
    def __init__(self):
        self.tamaño = random.randrange(10, 50)
        self.generacion = 0
        self.individuos = [Individuo() for _ in range(self.tamaño)]

    def evaluar(self):
        for ind in self.individuos:
            ind.evaluar_aptitud()

    def media_aptitud(self):
        return sum(ind.aptitud for ind in self.individuos) / len(self.individuos)

    def sobrevive(self):
        media = self.media_aptitud()
        self.individuos = [ind for ind in self.individuos if ind.aptitud >= media]

    def seleccionar(self):
        a, b = random.sample(self.individuos, 2)
        return a if a.aptitud > b.aptitud else b

    def evolucionar(self):
        self.evaluar()
        self.sobrevive()
        nueva = []
        while len(nueva) < self.tamaño:
            padre1 = self.seleccionar()
            padre2 = self.seleccionar()
            hijo = padre1.cruzar(padre2)
            hijo.mutar()
            nueva.append(hijo)
        self.individuos = nueva
        self.generacion += 1

    def mejor(self):
        return max(self.individuos, key=lambda i: i.aptitud)

    def peor(self):
        return min(self.individuos, key=lambda i: i.aptitud)


    def mostrar(self):
        print("Generación:", self.generacion)
        print("Tamaño:    ", self.tamaño)
        print("Media:     ", round(self.media_aptitud(), 2))
        print("Mejor individuo:")
        self.mejor().mostrar()
        print("Peor individuo:")
        self.peor().mostrar()


p = Poblacion()
p.evaluar()
print("--- Antes de evolucionar ---")
p.mostrar()

for _ in range(10):
    p.evolucionar()

print("\n--- Después de 10 generaciones ---")
p.evaluar()
p.mostrar()
```
