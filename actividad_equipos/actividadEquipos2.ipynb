```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv("data/CincinnatiPaintCompany.csv")
df.head()
```

```python
y = df[['Commissions']]
y
```

```python
X = df[['Driven','Calls']]
X
```

```python
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)
y_pred
```

```python
mse = mean_squared_error(y,y_pred)
r2 = r2_score(y, y_pred)
print("la mean square error:",mse)
print("el varianza",r2)
```

```python
model.intercept_ #cuando los valores sean 0 0, cual sera el valor predecido
```

```python
print('coeficientes: ',model.coef_) #por cada llamada las comisiones aumentan en 0.632 y por cada milla en 0.015
print("error cuadratico medio",mse) #nos dice que en promedio el error del modelo es bajo
print("varianza", r2) #nos indica que el modelo tiene un buen ajuste
```

```python
Y = model.predict([[0,0]])
print("las comisiones generadas serian:", Y)
```
