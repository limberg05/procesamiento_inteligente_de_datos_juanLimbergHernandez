```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Importe de Datos
df = pd.read_csv("data/Coleccionista.csv")
df.head()
```

```python
X = df[['Bidders', 'Age']]
X
```

```python
y =df[['Price']]
y
```

```python
model = LinearRegression()
model.fit(X,y)
y_pred = model.predict(X)

mse = mean_squared_error(y,y_pred)
r2 = r2_score(y,y_pred)
```

```python
model.intercept_
```

```python
print('Coefcientes: ', model.coef_)
print('Error cuadratico de la media', mse)
print('Variaza: ', r2)
```

```python
Y = model.predict([[10,1000]])
print('El precio de la pintura es: ',Y)
```

# Relacion de los datos obtenidos

La regresión muestra que la antigüedad incrementa el precio de la pintura, mientras que el número de postores tiene una relación inversa, por lo que mientras mas incrementa el numero de postores el precio disminuye, posiblemente porque las obras más valiosas atraen a menos compradores especializados.
