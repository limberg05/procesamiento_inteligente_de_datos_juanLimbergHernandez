### Regresión Lineal Múltiple - High Point Sofa and Chair
1. Determinar las variables independientes
2. Determinar la variable dependiente
3. Obtener el valor de interseccion
4. Estimar el corto de calefaccion para una Temperatura externa media de 30 grados, 5 pulgadas de aislamiento y un calentador con antiguedad de 10 años.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Importar datos
df = pd.read_csv('data/HighPointSofaandChair.csv')
df.head()
```

# Determinar las variables independientes
x1,x2,y3

```python
X = df[['Aptitude', 'Union']]
X
```

```python
# Determinar las variables Dependientes
y = df[['Performance']]
y
```

```python
# Ajustar el modelo de regresión multiple
model = LinearRegression()
model.fit(X,y) # entrenando el modelo
y_pred = model.predict(X)
#y_pred

# Métricas estadísticas
mse = mean_squared_error(y, y_pred) # Error de la media cuadrática
r2 = r2_score(y,y_pred) # varianza
```

# Calcular el valor de Interseccion
Cuando la aptitude y union son 0

```python
model.intercept_
```

# Coeficiente
En cuanto aumenta el performance por cada punto de aptitude y union

# Error cuadrático medio
Mientras más bajo mejor, el modelo se equivoca por 15.13 puntos en promedio

# Varianza 
El modelo explica aproximadamente el 53.3% de la variación del desempeño laboral (0% el modelo no explica nada)

```python
print("Coeficientes: ", model.coef_)
print("Error cuadrático medio",mse)
print("Varianza: ", r2)
```

```python
Y = model.predict([[6,0]])
print("El costo de calefaccion estimado es: ", Y[0])
```

Los resultados de la regresión muestran que existe una relación positiva entre las variables independientes y el desempeño laboral: por cada punto adicional en la calificación del examen de aptitud (Aptitude), el desempeño laboral (Performance) aumenta en promedio aproximadamente 5.22 puntos, manteniendo constante la pertenencia al sindicato.
