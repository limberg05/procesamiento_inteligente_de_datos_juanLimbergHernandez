```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("data/habitaciones.csv")
```

```python
X_train = np.array(data["Habitaciones_ocupadas"]).reshape(-1,1)
y_train = np.array(data["Ingreso"])
```

```python
r = linear_model.LinearRegression()
r.fit(X_train, y_train)
y_pred = r.predict(X_train)

```

```python
print("coeficiente:", r.coef_)
print("interseccion:", r.intercept_)
print("Error cuadratico medio:", mean_squared_error(y_train, y_pred))
print("R2:", r2_score(y_train, y_pred))
```

```python
x = X_train.ravel()
plt.figure(figsize=(8,5))
plt.scatter(x, y_train, color="steelblue", s=80, label="Datos")
plt.plot(x, y_pred, color="red", linewidth=3, label="Recta de regresion")
plt.xlabel("Habitaciones ocupadas")
plt.ylabel("Ingreso del restaurante")
plt.title("correlacion lineal simple")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
